```python
yf.pdr_override()

start_date = '2016-01-01'
end_date = '2021-5-5'
stock_data = data.get_data_yahoo('INTC', start_date, end_date)
stock_data.head()
```

```python
stock_data['Close'].plot(figsize=(12, 6), grid=True) # 종가 그리기
```

```python
stock_data_trunc=stock_data[:'2020-11-08'] # ~ 2020-11-08
stock_data_trunc.tail()
```

```python
stock_data.tail()
```

```python
stock_data_trunc.index
```

```python
df = pd.DataFrame({'ds':stock_data_trunc.index, 'y':stock_data_trunc['Close']}) # index = 날짜값
```

```python
df
```

```python
df.reset_index(inplace=True) # index reset, inplace=True 적용
```

```python
del df['Date']
```

```python
df
```

```python
#1년후, 2021-11-08까지 예측하기


m = Prophet(daily_seasonality=True)
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=365) # periods = 향후 x일 dataframe make / 1년
```

```python
future.tail() # 2021-11-06
```

```python
forecast = m.predict(future) # 2021-11-06까지 예측하기
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail() # 예측한 값
```

```python
m.plot(forecast); # balck = original_data
```

```python
m.plot_components(forecast)
```

```python
#시각화

plt.figure(figsize=(12,6))
plt.plot(stock_data.index, stock_data['Close'], label='real')
plt.plot(forecast['ds'], forecast['yhat'], label='forecast')
plt.grid()
plt.legend()
plt.show()
```
