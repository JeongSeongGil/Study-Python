```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
url_base = 'http://www.chicagomag.com'
url = 'http://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'

html = urlopen(url)
soup = BeautifulSoup(html, "html.parser")

soup
```

```python
from urllib.parse import urljoin
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()
    
    main_menu.append(re.split(('\r\n|\n'), tmp_string)[0])
    cafe_name.append(re.split(('\r\n|\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
print(rank[:5], main_menu[:5], cafe_name[:5])
```

```python
import pandas as pd
```

```python
data = {'Rank':rank, 'Menu':main_menu, 'Cafe':cafe_name, 'URL':url_add}

df = pd.DataFrame(data)

df.head()
```

```python
df = pd.DataFrame(data, columns=["Rank", "Cafe", "Menu", "URL"])

df.head(10)
```

```python
df.to_csv('../data/03. best_sandwiches_list_chicago.csv', sep=',', 
          encoding='UTF-8')
```

```python
from tqdm import tqdm_notebook

price = []
address = []

for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
df['Price'] = price
df['Address'] = address

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace=True)
df.head()
```

```python
df.to_csv('../data/03. best_sandwiches_list_chicago2.csv', sep=',', 
          encoding='UTF-8')
```

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('../data/03. best_sandwiches_list_chicago2.csv', index_col=0)
df.head(5)
```

```python
gmaps_key = "****************************************"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n] + ', ' + 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat) , len(lat)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start = 11)

folium.Marker([df['lat'].mean(), df['lng'].mean()], popup = 'center').add_to(mapping)

mapping
```

```python
mapping = folium.Map(location = [df['lat'].mean(), df['lng'].mean()],
                     tiles = 'Stamen Toner', zoom_start = 11)
for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], popup = df['Cafe'][n]).add_to(mapping)

mapping
```
